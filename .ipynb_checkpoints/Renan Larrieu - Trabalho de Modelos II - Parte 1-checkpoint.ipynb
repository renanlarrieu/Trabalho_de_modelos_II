# 2) Exemplo sistemas com parâmetros constantes e com parâmetros variando no tempo

Em uma queda livre ideal, isto é, desprezando a resistência do ar que gera força de arrasto no corpo, tem-se que a aceleração do corpo é constante no tempo e igual a aceleração da gravidade, com isso:


```python
t2=np.linspace(0,6,1000)
```

## $$t'=10s$$

```python
tempo_constante = 10 #segundos
```

```python
g=9.8
```

```python
a_corpo=g
```

O sistema de referência adotado faz com que a posição inicial do corpo seja nula:

```python
So=0
```

Assume-se, pelo objeto estar inicialmente em repouso, que sua velocidade inicial então é nula

```python
vo=0
```

```python
v=vo+a_corpo*t2
```

```python
S=So+vo*t2+(a_corpo+t2**2)/2
```

```python
g_array = np.full((1000), g)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2,sharex=True)
fig.suptitle('Parâmetros variáveis no tempo')
ax1.plot(t2, S,linewidth=3.)
ax1.grid(True)
ax1.set(xlabel='Tempo - s', ylabel='Posição - m')
ax2.plot(t2,v,color='r',linewidth=3.)
ax2.set(xlabel='Tempo - s', ylabel='Velocidade - m/s')
ax2.grid(True)


```

```python
plt.title('Parâmetro invariável no tempo')
plt.plot(t2,g_array,color='g',linewidth=3.)
plt.xlabel('Tempo - s')
plt.ylabel('Aceleração - m/s²')
plt.grid(True)
```

Observa-se que o sistema que tem aceleração constante, produziu queda livre com velocidade linear e posição quadrática

### Para $t'$ igual a uma quantidade finita, tem-se que

## $$ g(t+t')=g(t)=cte $$

## $$ s(t+t')\neq s(t)\neq cte $$

```python
v2=vo+a_corpo*(t2+tempo_constante)
```

```python
S2=So+vo*t2+(a_corpo+(t2+tempo_constante)**2)/2
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2,sharex=True)
fig.suptitle("Parâmetros variáveis s(t+t') e v(t+t')")
ax1.plot(t2+tempo_constante, S2,linewidth=3.,color='g')
ax1.grid(True)
ax1.set(xlabel='Tempo - s', ylabel='Posição - m')
ax2.plot(t2+tempo_constante,v2,color='orange',linewidth=3.)
ax2.set(xlabel='Tempo - s', ylabel='Velocidade - m/s')
ax2.grid(True)
```

Verifica-se, portanto, que a posição e velocidade tiveram suas curvas mudadas, e expõe-se os pontos da curva abaixo:

Para a constante de tempo "tempo_constante"

```python
t2[0]==(t2[0]+tempo_constante)
```

Tem-se:

```python
v2[0]==v[0]
```

```python
S[0]==S2[0]
```

Portanto, prova-se que a velocidade e posição são variantes no tempo

## 3.2) Sistema contínuo no tempo

Uma função de transferência de energia qualquer é um sistema contínuo no tempo

## $$H=\frac{3s^2+2s+3}{s^2+s+3}$$

```python
H=co.tf([3,2,3],[1,1,3])
```

```python
t = np.linspace(0, 10, 10000)  #ignora-se o fato do intervalo ser dividido. Entende-se que o intervalo de tempo entre dois pontos é infinito
t1,y1 = co.step_response(H,t)
plt.title("Step Response")
plt.xlabel('Time - s')
plt.plot(t1,y1,color='r',linewidth=3.)
plt.grid(True)
```

Com isso, observa-se um sistema que a partir da entrada contínua, responde de forma contínua de modo a se estabilizar com amplitude igual a 1.0 como pode ser visto no gráfico acima.

# 4) Sistemas estáveis e instáveis

## 4.1) Sistema estável

No exemplo abaixo, determinar-se-á a estabilidade em função do parâmetro A.

![circuito_estabilidade1.png](circuito_estabilidade1.png)

Para plotar a saída das duas condições, precisa-se determinar a função de transferência de energia do circuito.

## $$H(s)=\frac{sL}{sL+A+R+s^2CLR}$$

Após resolver o circuito apresentado e encontrar a função de transferência acima, pode-se inferir que o circuito é estável para qualquer valor da constante A, uma vez que tal constante não multiplica o termo de frequência complexa 's', com isso os Polos de H(s) são:

## $$s=\frac{-L \pm\sqrt{L^{2}-4LCR(A+R)}}{2LCR}$$

Observa-se acima que a parte real não depende da constante A

```python
R=3
L=3
C=1/36
A=2
```

```python
import cmath
import math
```

```python
s1=(-L+np.sqrt(L**2-4*L*C*R*(A+R)))/2*L*C*R
```

```python
A=2
```

```python
H1=co.tf([L,A],[L*C*R,L,A+R])
```

```python
H1
```

## $$H(s)=\frac{3s+2}{\frac{1}{4}s^2+3s+5}$$

```python
t = np.linspace(0, 10, 10000)  
t2,y2 = co.step_response(H1,t)
plt.title("Step Response")
plt.xlabel('Time - s')
plt.plot(t2,y2,color='r',linewidth=3.)
plt.grid(True)
```

```python
t = np.linspace(0, 10, 10000)  
t2,y2 = co.impulse_response(H1,t)
plt.title("Impulse Response")
plt.xlabel('Time - s')
plt.plot(t2,y2,color='r',linewidth=3.)
plt.grid(True)
```

Portanto, tem-se que tanto para a resposta a um degrau quanto a um impulso, o sistema converge para um valor.

## 4.1) Sistema instável

Determina-se um sistema arbitrário instável para um determinado valor da constante $g_{m}$

```python
gm=3
```

```python
G=co.tf([3,0,3],[4,(2-gm),3]) #sistema instável
```

```python
G
```

## $$G(s)=\frac{3s^2+3}{4s^2-s+3}$$

```python
t = np.linspace(0, 50, 10000)  
t2,y2 = co.step_response(G,t)
plt.title("Step Response")
plt.xlabel('Time - s')
plt.plot(t2,y2,color='b',linewidth=3.)
plt.grid(True)
```

```python
t = np.linspace(0, 50, 10000)  
t2,y2 = co.impulse_response(G,t)
plt.title("Step Response")
plt.xlabel('Time - s')
plt.plot(t2,y2,color='b',linewidth=3.)
plt.grid(True)
```

Observa-se claramente que o sistema é instável para $g_{m} = 3$, pois a amplitude aumenta de maneira oscilatória a medida que o tempo passa.


Para um intervalo de tempo finito igual ao período de oscilação da resposta analisada:

## $$T=t'$$

Tem-se que o sistema $G(s)$ não é instável pois

## $$f(t+2T)>f(t+T)>f(t)$$

Dessa forma, afirma-se que o sistema **diverge**, e com isso, pode ser classificado como instável

## 5.2) Sistema Digital

Além disso, sabe-se que existem sinais digitais envolvidos no protocolo de comunicação, e estes por si só seriam um excelente exemplo para compreender o que é pedido. No entanto, a fim de simplificar a modelagem, imagina-se que o arduíno queira também indicar em tempo real para os operantes do teste quando ocorre o valor máximo de empuxo no intervalo especificado acima.

Dessa forma, tem-se um sistema digital, em que há troca de estado _LOW_ para estado _HIGH_, bem como é mostrado na modelagem abaixo

```python
t=[]
for i in range(0,len(y),1):
    t.append(i)
valores=[]
s=0
for i in range(0,len(y),1):
    if (y[i]==max(y)):
        s=1
        valores.append(s)
    else:
        s=0
        valores.append(s)
```

```python
plt.title('Acionamento de LED a partir de pico de Empuxo',fontsize=60)
plt.ylabel('Estado Lógico',fontsize=50)
plt.step(t,valores,linewidth=6.,color='r')
plt.tick_params(labelsize=50)
plt.grid(True)
```

Com isso, a entrada de tensão do led é digital, assim como a saída do sistema modelado, o que caracteriza o conjunto como um sistema **digital**
